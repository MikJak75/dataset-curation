```python
import json
import os
```

```python
dataset_location = "../../benchmark_image_datasets/imagenet-1k/"
```

```python

class_names = ["Rottweiler", "hammerhead, hammerhead shark",]

file = open('imagenet/classes.json', 'r')

class_dict = json.load(file)

for c in class_names :
    keys = [key for key, value in class_dict.items() if value == c]
    key = keys[0]
    path = dataset_location + "train/" + key + "/"
    file_count = len([f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))])
    print(file_count)
```
